# Sampling fron a distibution


```python
import numpy as np
import matplotlib.pyplot as plt

sample = 100

mean0 = [0,0]
mean1 = [-1,1]

cov0 = [[1,0],[0,1]]

Data0 = np.random.multivariate_normal(mean0, cov0, sample)
Data1 = np.random.multivariate_normal(mean1,cov0,sample)

plt.figure()
plt.plot(*Data0.T,'.',label = 'Data0')
plt.plot(*Data1.T,"*",label="Data1")
plt.axis('scaled')
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

sample = 100

mean0 = [0,0]

cov0 = [[1,0],[0,1]]
cov2 = [[1,0.5],[0.5,1]]
Data0 = np.random.multivariate_normal(mean0, cov0, sample)
Data2 = np.random.multivariate_normal(mean0,cov2,sample)

plt.figure()
plt.plot(*Data0.T,'.',label = 'Data0')
plt.plot(*Data2.T,"*",label="Data2")
plt.axis('scaled')


```


If the covariance matrix is changed to
$$
\begin{bmatrix}
1&-0.5\\
-0.5&1\\
\end{bmatrix}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

sample = 100

mean0 = [0,0]

cov0 = [[1,0],[0,1]]
cov3 = [[1,-0.5],[-0.5,1]]
Data0 = np.random.multivariate_normal(mean0, cov0, sample)
Data3 = np.random.multivariate_normal(mean0,cov3,sample)

plt.figure()
plt.plot(*Data0.T,'.',label = 'Data0')
plt.plot(*Data3.T,"*",label="Data3")
plt.axis('scaled')
plt.legend()
```
