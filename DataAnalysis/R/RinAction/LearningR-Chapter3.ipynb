## 一个绘图的实例

```r
dose <- c(20,30,40,45,60)

drugA <- c(16,20,27,40,60)

drugB <- c(15,18,25,31,40)

plot(dose,drugA,type = "b")
```

## 图形参数

```r
opar <- par(no.readonly = TRUE)

par(lty = 2, pch = 17)

plot(dose, drugA, type="b")

par(opar)
```

### 符号和线条

+ pch 指定符号

+ cex 指定符号大小，表示绘图符号相对于默认值的大小的缩放倍数

+ lty 指定线条类型

+ lwd 指定线条宽度

```r
plot(dose,drugA,type = "b",lty=3, lwd = 3, pch = 15, cex = 2)
```

### 颜色的指定

+ col 默认的绘图颜色

+ col.axis 坐标轴刻度文字的颜色

+ col.lab 坐标轴标签的颜色

+ col.main 标题颜色

+ col.sub 副标题颜色

+ fg 图形的前景色

+ bg 图形的背景色

可以通过指定颜色下标、颜色名称、十六进制的颜色值、RGB值和HSV值来指定颜色

推荐使用 **RColorBrewer**

```r
library(RColorBrewer)

n <- 7

mycolors <- brewer.pal(n,"Set1")

barplot(rep(1,n),col=mycolors)

n <- 10

mycolors <- rainbow(n)

pie(rep(1,n),labels = mycolors,col=mycolors)

mygrays <- gray(1:n/n)

pie(rep(1,n),labels = mygrays,col=mygrays)
```

### 文本属性

+ cex 指定大小

+ font 指定字体

### 图形尺寸和边界尺寸

+ pin 图形的宽和高

+ mai/mar

数值向量表示的边界大小，“下左上右”

### 添加文本、自定义坐标轴和图例

```r
par(lwd = 2,cex=1.5,font.lab=2)

plot(dose,drugB,type="b",
    
col = "red",lty = 2, pch = 2, lwd = 2,
    
main = "Clinical Trials for Drug A",
    
sub = "This is data",
    
xlab = "Dosage",ylab="Drug Response",
    
xlim = c(0,60),ylim = c(0,70))

abline(h=c(20),v=c(30,50))

lines(dose,drugA,type="b",pch=17,lty=2,col="blue")

legend("topleft",title="Drug Type",c("A","B"),bcol=c("red","blue"))
```

### 图形的组合

par()

layout()

```r
attach(mtcars)

opar <- par(no.readonly = TRUE)

par(mfrow = c(2,2))

plot(wt,mpg,main="Scatterplot of wt vs mpg")

plot(wt,disp,main = "Scatterplot of wt vs disp")

hist(wt,main="Histogram of wt")

boxplot(wt,main = "Boxplot of wt")

par(opar)

detach(mtcars)
```

```r
attach(mtcars)

opar <- par(no.readonly = TRUE)

par(mfrow=c(3,1))

hist(wt)

hist(mpg)

hist(disp)

par(opar)

detach(mtcars)
```

```r
attach(mtcars)

layout(matrix(c(1,1,2,3),2,2,byrow=TRUE))

hist(wt)

hist(mpg)

hist(disp)

detach(mtcars)
```
