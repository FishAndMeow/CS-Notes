## Creating a Scatter Plot

```r
plot(mtcars$wt,mtcars$mpg)
```

```r
library(ggplot2)
ggplot(mtcars,aes(x = wt, y = mpg))+geom_point()
```

The first part, ggplot() tells it to create a plot object, and the second part, geom_point() tells it to add a layer of points to the plot.

The usual way to use ggplot() is to pass it a data frame (mtcars) and then
tell it which columns to use for the x and y values. If you want to pass it two
vectors for x and y values, you can use data = NULL, and then pass it the
vectors. Keep in mind that ggplot2 is designed to work with data frames as
the data source, not individual vectors, and that using it this way will only
allow you to use a limited part of its capabilities:

```r
ggplot(data = NULL, aes(x = mtcars$wt, y = mtcars$mpg)) +
geom_point()
```

## Creating a Box Plot

```r
plot(ToothGrowth$supp,ToothGrowth$len)
```

If the two vectors are in the same data frame, you can also use the boxplot()
function with formula syntax

```r
# Formula syntax
boxplot(len ~ supp, data = ToothGrowth)
# Put interaction of two variables on x-axis
boxplot(len ~ supp + dose, data = ToothGrowth)
```

```r
ggplot(ToothGrowth,aes(x=supp,y=len))+
geom_boxplot()
```

It’s also possible to make box plots for multiple variables, by combining the
variables with interaction()

```r
ggplot(ToothGrowth, aes(x = interaction(supp, dose), y = len)) +
geom_boxplot()
```

## Ploting a Function Curve

```r
curve(x^3 - 5*x,from = -4 ,to = 4 )
```

Using add =
TRUE will add a curve to the previously created plot

```r
# Plot a user-defined function
myfun <- function(xvar) {
1 / (1 + exp(-xvar + 10))
} 
curve(myfun(x), from = 0, to = 20)
# Add a line:
curve(1 - myfun(x), add = TRUE, col = "red")
```

```r
ggplot(data.frame(x=c(0,20)),aes(x=x))+
stat_function(fun=myfun,geom='line')
```
